### Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

### Importing Dataset

```python
df = pd.read_csv("shopping_trends.csv")
df.head(5)
```

### Data Visualization 

```python
sns.histplot(x='Review Rating', data=df, )   #histogram
plt.show()
```

```python
sns.boxplot( x="Age", y='Category', data=df)  #boxplot  
plt.show() 
```

```python
sns.scatterplot( x="Age", y='Category', data=df,
                hue='Gender', size='Previous Purchases')  #scatterplot
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

```python
sns.pairplot(df, hue='Gender', height=2)  #pairplot
```

```python
df.hist(bins=50,figsize=(15,15)) #histogram
```

```python
np.random.seed(0)
sns.set()
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data, vmin=0, vmax=1)  #heatmap
plt.show()
```
